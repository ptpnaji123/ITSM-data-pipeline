```python
import pandas as pd

# Define file path
file_path = r"D:\ITSM\Sample_Data_file_for_Analysis_Jan25.csv"

# Try using ISO-8859-1 encoding
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Display the first 5 rows
df.head()
```

```python
print(df.columns)
```

```python


# Rename columns based on correct mapping
df.rename(columns={
    "inc_number": "Ticket ID",
    "inc_category": "Category",
    "inc_business_service": "Sub-Category",
    "inc_priority": "Priority",
    "inc_sys_created_on": "Created Date",
    "inc_resolved_at": "Resolved Date",
    "inc_state": "Status",
    "inc_assignment_group": "Assigned Group",
    "inc_assigned_to": "Technician",
    "inc_close_notes": "Customer Impact"
}, inplace=True)

# Display updated column names
print("Updated Column Names:", df.columns)
```

```python
# Convert dates to datetime format
df["Created Date"] = pd.to_datetime(df["Created Date"], format="%d-%m-%Y %H:%M", errors="coerce")
df["Resolved Date"] = pd.to_datetime(df["Resolved Date"], format="%d-%m-%Y %H:%M", errors="coerce")

# Display first few rows to check conversion
df[["Created Date", "Resolved Date"]].head()
```

```python
# Find and count duplicate Ticket IDs
duplicate_ticket_ids = df["Ticket ID"].duplicated(keep=False)  # Marks all duplicates

# Display the duplicate records
df_duplicates = df[duplicate_ticket_ids]
print("Total Duplicate Rows Found:", len(df_duplicates))

# Show a sample of duplicate rows
df_duplicates[["Ticket ID"]].head()
```

```python
# Drop duplicate Ticket IDs, keeping the first occurrence
df.drop_duplicates(subset=["Ticket ID"], keep="first", inplace=True)

# Reset index after dropping duplicates
df.reset_index(drop=True, inplace=True)
```

```python
df = df[[
    "Ticket ID", "Category", "Sub-Category", "Priority", "Created Date", 
    "Resolved Date", "Status", "Assigned Group", "Technician", "Customer Impact"
]]
```

```python
df
```

```python
import numpy as np

# Convert "Pending" values in Resolved Date back to NaT (Not a Time)
df["Resolved Date"] = pd.to_datetime(df["Resolved Date"], errors="coerce")

# Calculate resolution time in hours
df["Resolution Time (Hrs)"] = (df["Resolved Date"] - df["Created Date"]).dt.total_seconds() / 3600

# Replace missing or negative values with 0
df["Resolution Time (Hrs)"] = df["Resolution Time (Hrs)"].apply(lambda x: max(x, 0) if pd.notnull(x) else 0)

# Verify the column is created
df[["Created Date", "Resolved Date", "Resolution Time (Hrs)"]].head()
```

```python
df
```

```python
# Display missing values before handling
print("Missing Values Before Handling:\n", df.isnull().sum())
```

```python
# Fill missing Resolved Date with Created Date to maintain the date format
df["Resolved Date"] = df["Resolved Date"].fillna(df["Created Date"])

# Fill missing Customer Impact with "Not Specified"
df["Customer Impact"] = df["Customer Impact"].fillna("Not Specified")
```

```python
df.dtypes
```

```python
# Group by Category and calculate the average resolution time
avg_resolution_category = df.groupby("Category")["Resolution Time (Hrs)"].mean().reset_index()

# Rename the column for clarity
avg_resolution_category.rename(columns={"Resolution Time (Hrs)": "Avg Resolution Time (Hrs)"}, inplace=True)

# Display results
print("Average Resolution Time per Category:")
print(avg_resolution_category.head())
```

```python
# Group by Priority and calculate the average resolution time
avg_resolution_priority = df.groupby("Priority")["Resolution Time (Hrs)"].mean().reset_index()

# Rename the column for clarity
avg_resolution_priority.rename(columns={"Resolution Time (Hrs)": "Avg Resolution Time (Hrs)"}, inplace=True)

# Display results
print("Average Resolution Time per Priority:")
print(avg_resolution_priority.head())
```

## Ticket_Closure_Rate_by_Assigned_Group

```python
# Count unique Assigned Groups
num_assigned_groups = df["Assigned Group"].nunique()

# Display result
print("Total Number of Assigned Groups:", num_assigned_groups)
```

```python
# Count total tickets per Assigned Group
total_tickets = df.groupby("Assigned Group")["Ticket ID"].count().reset_index()
total_tickets.rename(columns={"Ticket ID": "Total Tickets"}, inplace=True)

# Count closed tickets per Assigned Group
closed_tickets = df[df["Status"] == "Closed"].groupby("Assigned Group")["Ticket ID"].count().reset_index()
closed_tickets.rename(columns={"Ticket ID": "Closed Tickets"}, inplace=True)

# Merge total and closed ticket counts
ticket_closure_rate = total_tickets.merge(closed_tickets, on="Assigned Group", how="left")

# Fill NaN values (if a group has no closed tickets, set count to 0)
ticket_closure_rate["Closed Tickets"] = ticket_closure_rate["Closed Tickets"].fillna(0)

# Calculate closure rate percentage
ticket_closure_rate["Closure Rate (%)"] = (ticket_closure_rate["Closed Tickets"] / ticket_closure_rate["Total Tickets"]) * 100

# Display results
print(ticket_closure_rate)
```

```python
closure_rate_file = r"D:\ITSM\Ticket_Closure_Rate_by_Assigned_Group.csv"
ticket_closure_rate.to_csv(closure_rate_file, index=False)

print("Ticket closure rate data saved successfully at:", closure_rate_file)
```

## Monthly Ticket Summary table aggregating the number of tickets, average resolution time, and closure rate per month.

```python
import pandas as pd

# Group by Year and Month to compute metrics
monthly_summary = df.groupby(["Year", "Month"]).agg(
    Total_Tickets=("Ticket ID", "count"),  # Count total tickets per month
    Avg_Resolution_Time_Hrs=("Resolution Time (Hrs)", "mean"),  # Average resolution time
    Closed_Tickets=("Status", lambda x: (x == "Closed").sum())  # Count closed tickets
).reset_index()

# Calculate Closure Rate
monthly_summary["Closure Rate (%)"] = (monthly_summary["Closed_Tickets"] / monthly_summary["Total_Tickets"]) * 100

# Drop unnecessary column
monthly_summary.drop(columns=["Closed_Tickets"], inplace=True)

# Display the summary as a table
print(monthly_summary.to_string(index=False))
```

```python
# Save the summary table as CSV
monthly_summary_file = r"D:\ITSM\Monthly_Ticket_Summary.csv"
monthly_summary.to_csv(monthly_summary_file, index=False)

print("Monthly ticket summary saved successfully at:", monthly_summary_file)
```
